## Task 1 Iris Dataset Explanation

```python
# Step 1: Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Step 2: Load the Iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Step 3: Dataset shape and column names
print('Shape:', iris.shape)
print('Columns:', iris.columns.tolist())
```

```python
## step 4 Dataset info and summary
iris.info()
iris.describe()
```

```python
# Step 5: Scatter Plot  Sepal Length vs Width
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Sepal Length vs Width by Species')
plt.show()
```

```python
## step 6 Histograms for all numeric features
iris.hist(bins=20, figsize = (10,8), edgecolor = 'black')
plt.suptitle('Feature Distribution')
plt.tight_layout()
plt.show()
```

```python
## step 7 box plots for outlier directions
plt.figure(figsize = (12,6))
for i, col in enumerate(iris.columns[:-1]):
    plt.subplot(1, 4, i+1)
    sns.boxplot(y=iris[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```
